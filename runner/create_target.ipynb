```python
import numpy as np
import matplotlib.pyplot as plt

# Thiết lập thông số cho cụm
num_clusters = 10     # Số lượng cụm (nên là số chính phương để dễ chia lưới)
points_per_cluster = 5 # Số lượng điểm trong mỗi cụm
cluster_spread = 40    # Độ rộng của cụm
space_size = 1000        # Kích thước của không gian

# Tạo lưới cho các cụm
grid_size = int(np.sqrt(num_clusters))  # Kích thước của lưới
grid_spacing = space_size / (grid_size + 1) # Khoảng cách giữa các điểm trên lưới

# Tạo các tâm cụm phân phối đều
cluster_centers = []
for i in range(1, grid_size + 1):
    for j in range(1, grid_size + 1):
        center_x = i * grid_spacing
        center_y = j * grid_spacing
        cluster_centers.append((center_x, center_y))

cluster_centers = np.array(cluster_centers)

# Tạo điểm cho từng cụm
points = []
for center in cluster_centers:
    cluster_points = np.random.normal(loc=center, scale=cluster_spread, size=(points_per_cluster, 2))
    points.append(cluster_points)

# Chuyển danh sách thành mảng numpy
points = np.vstack(points)

# Vẽ biểu đồ
plt.figure(figsize=(10, 10))
plt.scatter(points[:, 0], points[:, 1], alpha=0.6, label='Points')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], color='red', marker='x', s=100, label='Cluster Centers')
plt.title('Uniformly Distributed Cluster Centers with Gaussian Distributed Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()
```
